# **Tunning the Model using the TPOT Regressor** 

```python
df_train.head()
```

```python
df_test.head()
```

```python
# Creating the dataframe for independent and dependent variables for test and train data
x_train= df_train.drop(['Product_Price'], axis=1)
x_test= df_test.drop(['Product_Price'], axis=1)
y_train= df_train['Product_Price']
y_test=  df_test['Product_Price'] 
```

```python
#pip install TPOT
```

```python
# Importing all the libraries used to create the tunned model

import numpy as np
import pandas as pd
from sklearn.ensemble import AdaBoostRegressor, ExtraTreesRegressor
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline, make_union
from sklearn.preprocessing import MaxAbsScaler, PolynomialFeatures, StandardScaler
from sklearn.svm import LinearSVR
from tpot.builtins import StackingEstimator
from xgboost import XGBRegressor
from tpot.export_utils import set_param_recursive
```

```python
# Creating the pipeline 
final_pipeline = make_pipeline(
    StackingEstimator(estimator=LinearSVR(C=15.0, dual=True, epsilon=0.01, loss="epsilon_insensitive", tol=0.1)),
    StandardScaler(),
    MaxAbsScaler(),
    StackingEstimator(estimator=AdaBoostRegressor(learning_rate=1.0, loss="square", n_estimators=100)),
    PolynomialFeatures(degree=2, include_bias=False, interaction_only=False),
    StackingEstimator(estimator=ExtraTreesRegressor(bootstrap=False, max_features=0.1, min_samples_leaf=2, min_samples_split=11, n_estimators=100)),
    StackingEstimator(estimator=XGBRegressor(learning_rate=0.001, max_depth=3, min_child_weight=6, n_estimators=100, n_jobs=1, objective="reg:squarederror", subsample=0.3, verbosity=0)),
    StackingEstimator(estimator=XGBRegressor(learning_rate=0.001, max_depth=5, min_child_weight=3, n_estimators=100, n_jobs=1, objective="reg:squarederror", subsample=0.45, verbosity=0)),
    ExtraTreesRegressor(bootstrap=False, max_features=0.9000000000000001, min_samples_leaf=16, min_samples_split=7, n_estimators=100)
)
```

```python
# Fix random state for all the steps in exported pipeline
set_param_recursive(final_pipeline.steps, 'random_state', 42)

final_pipeline.fit(x_train.values, y_train.values)
```

```python
# prediction on test data set 
y_test_pred = final_pipeline.predict(x_test)

# prediction on train data set 
y_train_pred = final_pipeline.predict(x_train)
```

```python
#Finding error test 
mse_test = mean_squared_error(y_test, y_test_pred)
test_rmse = math.sqrt(mse_test)
print(test_rmse)
```

```python
#R2 of test
from sklearn.metrics import r2_score
r2_model_test = r2_score(y_test, y_test_pred)
r2_model_test
```

```python
#Finding error on train
mse_train = mean_squared_error(y_train, y_train_pred)
train_rmse = math.sqrt(mse_train)
print(train_rmse)
```

```python
#R2 of train
r2_model_train = r2_score(y_train, y_train_pred)
r2_model_train
```
